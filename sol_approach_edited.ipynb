```python
url="https://s3.us-east-2.amazonaws.com/datafaculty/bootcamp/membership.csv"
```

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv(url)
```

```python
data.head()
```

```python
data=data.drop(['MEMBERSHIP_NUMBER','AGENT_CODE'],axis=1)
```

```python
data.head()
```

```python
data.columns
```

```python
data=data.drop(['START_DATE (YYYYMMDD)','END_DATE  (YYYYMMDD)'],axis=1)
```

```python
data.columns
```

```python
data.isnull().sum()
```

```python
data['MEMBER_MARITAL_STATUS'].value_counts()
```

```python
data['MEMBER_MARITAL_STATUS']=data['MEMBER_MARITAL_STATUS'].fillna('Missing')
```

```python
data['MEMBER_GENDER']=data['MEMBER_GENDER'].fillna("Missing")
```

```python
data['MEMBER_OCCUPATION_CD'].unique()
```

```python
data['MEMBER_OCCUPATION_CD'].value_counts()
```

```python
ind_m=data[data['MEMBER_OCCUPATION_CD'].isnull()].index
```

```python
data.shape
```

```python
data=data.drop(ind_m,axis=0)
```

```python
data.isnull().sum()
```

```python
data['MEMBER_ANNUAL_INCOME'].describe()
```

```python
data['MEMBER_ANNUAL_INCOME']=data['MEMBER_ANNUAL_INCOME'].fillna(5.499960e+05)
```

```python
data.isnull().sum()
```

```python
## Label encode categorical variables
import sklearn.preprocessing as preprocessing
```

```python
le=preprocessing.LabelEncoder()
```

```python
data['MEMBER_MARITAL_STATUS']=le.fit_transform(data['MEMBER_MARITAL_STATUS'])
```

```python
data
```

```python
data['MEMBERSHIP_STATUS'].replace('INFORCE',1,inplace=True)
```

```python
data['MEMBERSHIP_STATUS'].replace('CANCELLED',0,inplace=True)
```

```python
data.head(3)
```

```python
data=pd.get_dummies(data)
```

```python
data.columns
```

```python
data.head(2)
```

```python
data=data.drop(['MEMBER_GENDER_Missing','MEMBERSHIP_PACKAGE_TYPE-A','PAYMENT_MODE_ANNUAL','PAYMENT_MODE_SINGLE-PREMIUM',],axis=1)
```

```python
y=data['MEMBERSHIP_STATUS']
x=data.drop('MEMBERSHIP_STATUS',axis=1)
```

```python
y.head(3)
```

```python
x.head(3)
```

```python
import sklearn.model_selection as model_selection
```

```python
x_train,x_test,y_train,y_test=model_selection.train_test_split(x,y,test_size=0.2,random_state=200)
```

```python
x_train.head(4)
```

```python
y_train.head(3)

```

```python
import sklearn.ensemble as ensemble
```

```python
##ranfor=ensemble.RandomForestClassifier()
```

```python
ranfor=ensemble.RandomForestClassifier(n_estimators=100,oob_score=True,n_jobs=-1)
```

```python
ranfor.fit(x_train,y_train)
```

```python
ranfor.predict(x_test)
```

```python
ranfor.predict_proba(x_test)[:,1]
```

```python
for i in [100,200,300]:
    clf=ensemble.RandomForestClassifier(n_estimators=i,oob_score=True,n_jobs=-1)
    clf.fit(x_train,y_train)
    print("oob score for {} trees is {}".format(i,clf.oob_score_))
```

```python
for j in range(10,400,30):
    ranfor=ensemble.RandomForestClassifier(n_estimators=j,oob_score=True,n_jobs=-1)
    ranfor.fit(x_train,y_train)
    oob=ranfor.oob_score_
    print("oob score for {} trees is {}".format(j,ranfor.oob_score_))
```

```python
ranfor.feature_importances_
```

```python
ranfor.oob_score_
```

```python
pd.Series(ranfor.feature_importances_,index=x_train.columns).sort_values(ascending=False)
```

```python
ranfor.oob_decision_function_[:,1]
```

```python
pred_oob=ranfor.oob_decision_function_[:,1]
```

```python
import sklearn .metrics as metrics
```

```python
metrics.roc_auc_score(y_train,pred_oob)
```

```python
for j in [160,280,370]:
    ranfor=ensemble.RandomForestClassifier(n_estimators=j,oob_score=True,n_jobs=-1)
    ranfor.fit(x_train,y_train)
    pred_oob=ranfor.oob_decision_function_[:,1]
    auc=metrics.roc_auc_score(y_train,pred_oob)
    print("Auc for {} trees is {}".format(j,auc))
    
```
